```
from keras.models import Model
from keras.layers import Input, Dense, SimpleRNN, LSTM
import numpy as np
```

```
# first data set
h, e, l, o = 0, 1, 2, 3
x = np.array([[[h], [e], [l], [l]]], dtype=np.float32)
y = np.array([[o]], dtype=np.float32)

# second data set
#l, e, a, r, n, i, g = 0, 1, 2, 3, 4, 5, 6
#x = np.array([[[l], [e], [a], [r], [n], [i], [n]]], dtype=np.float32)
#y = np.array([[g]], dtype=np.float32)

print("x:", x, x.shape)
print("y:", y, y.shape)
```

```
model_type = 2 # 1: Regression, 2: Classification
```

```
x_input = Input(batch_shape=(None, x.shape[1], 1))
rnn1 = LSTM(4, return_sequences=True)(x_input)
rnn2 = LSTM(4)(rnn1)
if model_type == 1:
  x_output = Dense(1)(rnn2)
elif model_type == 2:
  x_output = Dense(4, activation="softmax")(rnn2)
```

```
model = Model(x_input, x_output)

if model_type == 1:
  model.compile(loss='mean_squared_error', optimizer='adam')
elif model_type == 2:
  model.compile(loss='sparse_categorical_crossentropy', optimizer='adam')

print(model.summary())
```

```
model.fit(x, y, epochs=1000, batch_size=1, verbose=0)
```

```
pred = model.predict(x, batch_size=1)
if model_type == 1:
  print(pred)
elif model_type == 2:
  print(np.round(pred, 3))
  idx = [np.argmax(r) for r in pred]
  print(idx)
```
