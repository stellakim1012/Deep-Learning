```python
# setup
import numpy as np
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, SimpleRNNCell,SimpleRNN, RNN
```

### Preparing dataset

```python
# One hot encoding for each char in 'hello'
h = [1, 0, 0, 0]
e = [0, 1, 0, 0]
l = [0, 0, 1, 0]
o = [0, 0, 0, 1]
```

```python
# One cell RNN input_dim (4) -> output_dim (2)
x_data = np.array([[h]], dtype=np.float32)
print(x_data.shape)

hidden_size = 2
cell = SimpleRNNCell(units=hidden_size) # creating SimpleRNNCell
rnn = RNN(cell, return_sequences=True, return_state=True)

outputs, states = rnn(x_data)
print("파라미터 개수:", rnn.count_params())
print(rnn.weights)

print('x_data: {}, shape: {}'.format(x_data, x_data.shape))
print('outputs: {}, shape: {}'.format(outputs, outputs.shape))
print('states: {}, shape: {}'.format(states, states.shape))
```

```python
# equivalent to above case
rnn = SimpleRNN(units=hidden_size, return_sequences=True,
               return_state=True) # layers.SimpleRNNCell + layers.RNN

outputs, states = rnn(x_data)
print("파라미터 개수:", rnn.count_params())
print(rnn.weights)

print('x_data: {}, shape: {}'.format(x_data, x_data.shape))
print('outputs: {}, shape: {}'.format(outputs, outputs.shape))
print('states: {}, shape: {}'.format(states, states.shape))
```

```python
# One cell RNN input_dim (4) -> output_dim (2). sequence: 5
x_data = np.array([[h, e, l, l, o]], dtype=np.float32)

hidden_size = 2
rnn = SimpleRNN(units=hidden_size, return_sequences=True, return_state=True)
#rnn = SimpleRNN(units=hidden_size, return_state=True, input_shape=(1, 4))
outputs, states = rnn(x_data)
print(rnn.weights)

print('x_data: {} shape: {} \n'.format(x_data, x_data.shape))
print('outputs: {} shape: {} \n'.format(outputs, outputs.shape))
print('states: {}, shape: {}'.format(states, states.shape))
```

```python
# One cell RNN input_dim (4) -> output_dim (2). sequence: 5, batch 3
# 3 batches 'hello', 'eolll', 'lleel'
x_data = np.array([[h, e, l, l, o],
                   [e, o, l, l, l],
                   [l, l, e, e, l]], dtype=np.float32)

hidden_size = 2
rnn = SimpleRNN(units=2, return_sequences=True, return_state=True)    
outputs, states = rnn(x_data)

print('x_data: {}, shape: {} \n'.format(x_data, x_data.shape))
print('outputs: {}, shape: {} \n'.format(outputs, outputs.shape))
print('states: {}, shape: {}'.format(states, states.shape))
```
